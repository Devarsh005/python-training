# while loop


### while loop iterate on list

```python
a_list =list(range(0,20))
index = 0
while index<len(a_list):
    print(a_list[index],end=' ')
    index+=1
```

```python
tup = tuple(range(2,20))
index =0
while index<len(tup):
    print(tup[index],end="  ")
    index+=1
```

### continue

```python
# write a number except 3
def number(n):
    for i in range(n):
        if i==3:
            continue
        else:
            print(i)

number(10)
```

```python
# write a number except 3 0r 5 or 7
def number(n):
    for i in range(n):
        if i  in [3,5,7]:
            continue
        else:
            print(i)

number(10)
```

### why we need a flag 

```python
outer_flag = False
while True:
    for num in range(100):
        if num>10:
            outer_flag=True
            break
        else:
            print(num,end="  ")
   
        
        
```

- in above example in if statement after break go on while statement and while statement is True so recent process start again and again 
- so in this condition we need to create a flag which is identify the process is done and break the loop


```python
outer_flag = False
while True:
    for num in range(100):
        if num>10:
            outer_flag=True
            break
        else:
            print(num,end="  ")
    if outer_flag:
        break
```

## raising an exception

```python
class getOutLopp(Exception):
    pass
while True:
    try :
        for num in range(100):
            if num >10:
                raise getOutLopp
            else:
                print(num,end=' ')
    except getOutLopp:
        break
```

## using a function in wrapping loops

```python
def myloop():
    while True:
        for num in range(100):
            if num>10:
                return
            else:
                print(num,end=" ")
myloop()
```

## using generator 

```python
def my_generator():
    while True:
        for i in range(100):
            if i>10:
                yield False
            else:
                yield i
for i in my_generator():
    if i is not False:
        print(i,end="  ")
    else:
        break
```
